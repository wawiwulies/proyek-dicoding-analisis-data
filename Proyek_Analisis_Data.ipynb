## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa kategori produk yang paling banyak terjual di e-commerse?

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title('Top 10 Most Sold Product Categories')
plt.xlabel('Number of Items Sold')
plt.ylabel('Product Category')
plt.show()
```

### Pertanyaan 2: Kapan terakhir pelanggan melakukan transaksi?

```python
# Alur dalam mendistribusikan transaksi tahun ke tahun

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.histplot(order_df['order_purchase_timestamp'], bins=50, kde=True)
plt.title('Distribution of Customer Transactions Over Time')
plt.xlabel('Transaction Date')
plt.ylabel('Frequency')
plt.show()
```
